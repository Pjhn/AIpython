```python
data = pd.Series([10, 11, 12, 13, 14, 15, 100, 101])
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
print(data[(data < lower_bound) | (data > upper_bound)])
```

```python
#점수 이상치 측정
Q1_score = df['점수'].quantile(0.25)
Q2_score = df['점수'].quantile(0.75)
IQR_score = Q2_score - Q1_score

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q2_score + 1.5 * IQR_score
```

```python
score_outliers = df[(df['점수'] < lower_bound_score) | (df['점수'] > upper_bound_score)]
print(score_outliers)
```

```python
#공부시간 이상치 측정
Q1_study = df['공부시간'].quantile(0.25)
Q2_study = df['공부시간'].quantile(0.75)
IQR_study = Q2_study - Q1_study

lower_bound_study = Q1_study - 1.5 * IQR_study
upper_bound_study = Q2_study + 1.5 * IQR_study
```

```python
#전체 이상치 제외 데이터(점수, 공부시간)
df_no_outliers = df[(df['점수'] >= lower_bound_score) & (df['점수'] <= upper_bound_score) & (df['공부시간'] >= lower_bound_study) & (df['공부시간'] <= upper_bound_study)]
```

```python
df_no_outliers
```

```python
#분반별 평균점수
df_clear= df_no_outliers.groupby('분반')['점수'].mean()
print(df_clear.round(2))
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_clear.index, df_clear.values)
plt.xlabel('분반')
plt.ylabel('평균 점수')
plt.title('분반별 평균 점수')
plt.grid(axis='y', alpha = 0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_no_outliers['공부시간'], df_no_outliers['점수'])
plt.title('공부시간과 점수 관계')
plt.xlabel('공부시간')
plt.ylabel('점수')
plt.grid(axis='both', alpha = 0.3)

plt.show()
```
