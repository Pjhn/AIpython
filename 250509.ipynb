```python
import matplotlib.pyplot as plt

month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rain = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize = (12, 6)) #plot()     

plt.plot(month, rain, color = 'g', marker = 'o',linestyle = ":", label = " ") #plt.plot(x,y 'go:')  color, marker, linestyle 
plt.title(" ")
plt.xlabel("")
plt.ylabel("(mm)")
plt.legend() # label   

for i, value in enumerate(rain):
  plt.text(i,value + 10, str(value), ha = 'center', va = 'bottom')
plt.show()
```
