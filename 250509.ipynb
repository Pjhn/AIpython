```python
import matplotlib.pyplot as plt

month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rain = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize = (12, 6)) #plot(꺾은선그래프)를 그리기 전에 미리 설정해야 함

plt.plot(month, rain, color = 'g', marker = 'o',linestyle = ":", label = "강수량 변화") #plt.plot(x,y 'go:') 축약형 color, marker, linestyle 순서
plt.title("월별 강수량")
plt.xlabel("월")
plt.ylabel("강수량(mm)")
plt.legend() #그래프의 label을 미리 설정해줘야 함

for i, value in enumerate(rain):
  plt.text(i,value + 10, str(value), ha = 'center', va = 'bottom')
plt.show()
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print("Shape:", arr.shape)    # (2, 3) 행과 열
print("Data Type:", arr.dtype)  # int64 데이터 타입
print("Size:", arr.size)      # 6 전체 요소의 개
print("Dimension:", arr.ndim) # 2 차원 수
```

```python
matrix = np.array([[1, 2], [3, 4], [5, 6]])

print(matrix[0, 1])      # 첫 번째 행, 두 번째 열
print(matrix[1:])        # 두 번째 행 이후 모두
```
