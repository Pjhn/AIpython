```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rain = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize = (12, 6)) #plot(꺾은선그래프)를 그리기 전에 미리 설정해야 함

plt.plot(month, rain, color = 'g', marker = 'o',linestyle = ":", label = "강수량 변화") #plt.plot(x,y 'go:') 축약형 color, marker, linestyle 순서
plt.title("월별 강수량")
plt.xlabel("월")
plt.ylabel("강수량(mm)")
plt.legend() #그래프의 label을 미리 설정해줘야 함

for i in range(len(month)):
    plt.annotate(month[i], (i,rain[i]), xytext = (5,-10), textcoords = 'offset points')

plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
jpn = [523, 529, 536, 532, 527]

# 그래프가 가로와 세로 크기를 각각 5인치와 3인치 크기 설정
plt.figure(figsize = (5,3))
# year 데이터를 x축으로, kor 데이터를 y축으로 하여 빨간색 점선으로 선 그래프를 그리고 범례에 'Korea'라고 표시
plt.plot(year, kor, 'r:', label = 'Korea' )#범례에 'Korea'라고 표시
plt.plot(year, usa, 'b:', label = 'USA' ) 
plt.plot(year, jpn, 'g-', label='Japan')
plt.ylim(400, 600)  # y축 범위 설정
# 그래프에 범례를 추가
plt.legend()
# 그래프 제목을 'Mathematics performance'로 설정합니다.
plt.title("Mathematics performance")
plt.xlabel('year')
plt.ylabel('score')
plt.show()
```
